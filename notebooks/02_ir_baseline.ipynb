```python
from rank_bm25 import BM25Okapi
from datasets import load_dataset
import numpy as np
```

```python
dataset = load_dataset("bigbio/bc5cdr")
```

```python
tokenized_docs = []

for doc in dataset["train"]:
    full_text = ""
    for passage in doc["passages"]:
        full_text += passage["text"] + " "
    tokenized_docs.append(full_text.split())
```

```python
bm25 = BM25Okapi(tokenized_docs)
```

```python
def search(query):
    tokenized_query = query.split()
    scores = bm25.get_scores(tokenized_query)
    top_5_indices = np.argsort(scores)[-5:][::-1]
    return top_5_indices
```

```python
results = search("chemical induced hypertension")
print(results)
```

```python
for result in results:
    print(dataset["train"][int(result)]["passages"][0]["text"])
```

```python
queries = [
    "ADHD",
    "Crohn's disease",
    "cancer treatment chemotherapy",
    "diabetes insulin medication",
    "cardiovascular disease drug treatment",
    "neurotoxicity chemical exposure",
    "kidney disease drug induced",
    "antibiotic resistance infection",
    "pain medication opioid effects"
]
```

```python
for query in queries:
    print(f"\nQuery: {query}")
    print("-" * 50)
    results = search(query)
    for result in results:
        print(dataset["train"][int(result)]["passages"][0]["text"])
```
